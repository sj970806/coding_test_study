# 문제 2
- https://leetcode.com/contest/weekly-contest-324/problems/smallest-value-after-replacing-with-sum-of-prime-factors/

```python
import math
class Solution:
    def smallestValue(self, n: int) -> int:
        if n == 4:
            return 4
        def primeFactor(n: int):
            tmp = []
            while n%2 == 0:    # 2로 먼저 나눔 (안나눠질때까지)
                tmp.append(2)
                n = n // 2

            for i in range(3, n+1, 2):  # 3부터 확인, 짝수는 나올 수가 없으니까 홀수만 검사
                if n % i == 0:
                    while n % i == 0:
                        tmp.append(i)
                        n = n // i

            return tmp

        ans = 0
        while True:
            res = primeFactor(n)
            if len(res) < 1:
                break
            elif len(res) == 1:
                ans = res[0]
                break
            n = sum(res)
            ans = n

        return ans
```
