# 문제 1
- 2 x n 타일링
- https://school.programmers.co.kr/learn/courses/30/lessons/12900

```문제 설명```
```
가로 길이가 2이고 세로의 길이가 1인 직사각형모양의 타일이 있습니다. 이 직사각형 타일을 이용하여 세로의 길이가 2이고 가로의 길이가 n인 바닥을 가득 채우려고 합니다. 타일을 채울 때는 다음과 같이 2가지 방법이 있습니다.

타일을 가로로 배치 하는 경우
타일을 세로로 배치 하는 경우
예를들어서 n이 7인 직사각형은 다음과 같이 채울 수 있습니다.
```

```
직사각형의 가로의 길이 n이 매개변수로 주어질 때, 이 직사각형을 채우는 방법의 수를 return 하는 solution 함수를 완성해주세요.
```
```제한사항```
```
가로의 길이 n은 60,000이하의 자연수 입니다.
경우의 수가 많아 질 수 있으므로, 경우의 수를 1,000,000,007으로 나눈 나머지를 return해주세요.
```
```입출력 예```
```
n	result
4	5
```
```입출력 예 설명```
```
입출력 예 #1
다음과 같이 5가지 방법이 있다.
```






````python
def solution(n):
    if n == 2:
        return 2
    elif n == 3:
        return 3
    else:
        cnt = 3
        n1 = 2
        n2 = 3 
        while True:
            n1,n2 = n2, n1+n2
            cnt += 1
            if cnt == n:
                return n2 % 1000000007
            
    
# from math import factorial
# def solution(n):    
#     answer = 0

#     max_cnt_2 = n//2
#     for cnt_2 in range(max_cnt_2+1):
#         cnt_1 = n - 2*cnt_2
#         res = factorial(cnt_1 + cnt_2) / (factorial(cnt_1) * factorial(cnt_2))   
#         # print(cnt_1,cnt_2, ':',int(res))
#         answer += int(res)
    
#     answer = answer % 1000000007
#     return answer
````

# 문제 2
- 정수 삼각형
- https://school.programmers.co.kr/learn/courses/30/lessons/43105

```문제 설명```

```

위와 같은 삼각형의 꼭대기에서 바닥까지 이어지는 경로 중, 거쳐간 숫자의 합이 가장 큰 경우를 찾아보려고 합니다. 아래 칸으로 이동할 때는 대각선 방향으로 한 칸 오른쪽 또는 왼쪽으로만 이동 가능합니다. 예를 들어 3에서는 그 아래칸의 8 또는 1로만 이동이 가능합니다.

삼각형의 정보가 담긴 배열 triangle이 매개변수로 주어질 때, 거쳐간 숫자의 최댓값을 return 하도록 solution 함수를 완성하세요.
```
```제한사항```
```
삼각형의 높이는 1 이상 500 이하입니다.
삼각형을 이루고 있는 숫자는 0 이상 9,999 이하의 정수입니다.
```
```입출력 예```

```
triangle	result
[[7], [3, 8], [8, 1, 0], [2, 7, 4, 4], [4, 5, 2, 6, 5]]	30
```

````python
def solution(triangle):
    top = triangle[0][0]
    for i,each in enumerate(triangle):
        cp = each[:]
        if i == 0:
            pass
        elif i == 1:
            for j in range(len(each)):
                each[j] += top
        else:
            for j in range(len(triangle[i-1])):
                each[j] = max(each[j], cp[j]+triangle[i-1][j])
                each[j+1] = max(each[j+1], cp[j+1]+triangle[i-1][j])
    
    answer = max(triangle[-1])
    return answer
````
